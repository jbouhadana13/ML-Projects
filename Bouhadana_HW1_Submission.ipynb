Problem 2: Data Preparation and Feature Selection for Machine Learning (25
points)
In this problem, you will continue to use the dataset from problem 1. We recommend you use
fit() and transfrom() functions (see tip #2) to work through the following set of questions:

```python
from pandas import read_csv
from numpy import set_printoptions
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.ensemble import ExtraTreesClassifier

# read and load the csv data file
filename = "/Users/jeremybouhadana/Downloads/myClassDataSet.csv"
dataframe = read_csv(filename)
array = dataframe.values
# separate array into input and output components
X = array [:,0:8]
Y = array [:,8]
```

Use StandardScaler class to standardize data; Print the first 7 rows of your transformed
data with a precision of 3 using set printoptions from Numpy. (5 points)

```python
#scaler object
scaler = StandardScaler()

#fit and transform the data
scaled_data = scaler.fit_transform(X) 
np.set_printoptions(precision=3)

#print first 7 rows
scaled_data[:7]
```

Use SelectKBest class and f classif() function to select the 6 best features in your data.
Print the scores with a precision of 3 using set printoptions from Numpy; Print the first 6
rows of selected features. (5 points)

```python
#Use SelectKBest class and f_classif() function to select the 6 best features
selector = SelectKBest(f_classif, k=6)
X_new = selector.fit_transform(X, Y)

#Print the scores with a precision of 3
np.set_printoptions(precision=3)
print(selector.scores_)

#Print the first 6 rows
print(X_new[:6, :])
```

Use RFE class with logisticRegression algorithm to select the top 6 features of your data.
Print number of features, selected features, and feature ranking using attributes n features ,
support and ranking , respectively. (6 points)

```python
#Create logistic regression model
model = LogisticRegression()

#Create  RFE object and fit it to the data
rfe = RFE(estimator=model, n_features_to_select=6)
fit = rfe.fit(X, Y)

#Print number of features
print("Number of features: %d" % fit.n_features_)

#Print selected features
print("Selected features: %s" % fit.support_)

#Print feature ranking
print("Feature ranking: %s" % fit.ranking_)
```

Use PCA class to reduce the dimension of your data to 2; Use explained variance ratio
attribute of PCA to get the percentage of variance explained by each of the selected components
(features). Print the results. (6 points)

```python
#Initialize PCA class
pca = PCA(n_components=2)

#Fit data from array to PCA
pca.fit(array)

#Get  explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_

#Print percentage of Variance for each component
print("Percentage of variance explained by each of the selected components:")
for i, component in enumerate(explained_variance_ratio):
    print(f"Component {i + 1}: {component * 100:.2f}%")
```

Use ExtraTreesClassifier classifier along with feature importances attribute to estimate the importance score for each attribute and print the results. (3 points)

```python
#Create ExtraTreesClassifier classifier object
clf = ExtraTreesClassifier()

#Fit the classifier to the data
clf.fit(X,Y)

#Get feature importances, sort them, and get indices
importances = clf.feature_importances_
sorted_idx = np.argsort(importances)[::-1]

#Print the feature names and importances
print("Importance score:")
for i, name in enumerate(dataframe):
    print("{}: {}".format(name, importances[sorted_idx[i-1]]))
```
